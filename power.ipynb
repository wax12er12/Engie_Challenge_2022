```python



def weeklydiff(list1):
   
    count = 0
    sum = 0

    weekly = []
    count = 0
    sum = 0
    for i in range(len(list1)):
        sum += list1[i]
        count += 1
    
        if (count == 168):
            weekly.append(sum/(168))
            sum, count = 0, 0


    diff = []
    for i in range(len(weekly)-1):
        if(weekly[i]!=0):
            average=((weekly[i]-weekly[i+1])/weekly[i])*100
            diff.append(average)
    return diff[-1]
```

```python


def rivaldiff(TE,TO,RE,RO):
    weeklyEC = []
    count = 0
    sum = 0
    for i in range(len(TE)):
        sum += TE[i]
        count += 1
    
        if (count == 168):
            weeklyEC.append(sum/(168))
            sum, count = 0, 0

    occ=[]
    count = 0
    sum = 0
    for i in range(len(TO)):
        sum += TO[i]
        count += 1
    
        if (count == 168):
            occ.append(sum/(168))
            sum, count = 0, 0
            Tlastweek = weeklyEC[-1]/occ[-1]
    weeklyEC = []
    count = 0
    sum = 0
    for i in range(len(RE)):
        sum += RE[i]
        count += 1
    
        if (count == 168):
            weeklyEC.append(sum/(168))
            sum, count = 0, 0

    occ=[]
    count = 0
    sum = 0
    for i in range(len(RO)):
        sum += TO[i]
        count += 1
    
        if (count == 168):
            occ.append(sum/(168))
            sum, count = 0, 0
            Rlastweek = weeklyEC[-1]/occ[-1]
            
    difference=((Tlastweek- Rlastweek)/(Tlastweek))*100

    return difference



```
